```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('reto_precios.csv', usecols=['lat','lon','days_on_site','bathrooms','parking_lots','num_bedrooms','m2','final_price'])

corr = df.corr()
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr,cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(df.columns),1)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(df.columns)
ax.set_yticklabels(df.columns)
plt.show()
```
